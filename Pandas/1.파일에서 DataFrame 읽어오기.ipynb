### text 파일
- txt 파일도 read_csv로 읽어올 수 있음

```python
# ','로 구분되어 있는 경우 별도의 옵션 없이 사용가능
df = pd.read_csv("data/friend_list.txt")
```

```python
df
```

- ','가 아닌 다른 문자가 구분자로 쓰인 경우
  - delimiter 옵션으로 구분자 지정 가능

```python
df = pd.read_csv("data/friend_list_tab.txt")
```

```python
df
```

```python
df = pd.read_csv("data/friend_list_tab.txt", delimiter = '\t')
```

```python
df
```
