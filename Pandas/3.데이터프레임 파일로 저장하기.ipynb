# DataFrame 파일로 저장하기

```python
import pandas as pd
```

```python
friends = [
    {'name': 'John', 'age': 20, 'job': 'student'},
    {'name': 'Jenny', 'age': 30, 'job': None},
    {'name': 'Nate', 'age': 30, 'job': 'teacher'}
]
df = pd.DataFrame(friends)
df = df[['name', 'age', 'job']]
```

```python
df.head()
```

## DataFrame csv 파일로 저장하기

```python
df.to_csv('output/friends.csv')
```

```python
# index default = True, header default = True
df.to_csv('output/friends.csv', index = True, header = True)	# 위 결과와 동일
```

```python
help(df.to_csv)
```

```python
df.to_csv('output/friends_without_index.csv', index = False)	# index 없이 저장하기
```

```python
df.to_csv('output/friends_without_header.csv', header = False)	# header 없이 저장하기
```

```python
# 값이 없는 경우 내용 채우기 na_rep default = ''
df.to_csv('output/friends_na_rep.csv', na_rep = '-')
```
