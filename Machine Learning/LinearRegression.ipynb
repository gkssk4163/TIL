## Linear Regression (선형 회귀)

선형 회귀(Linear Regression)는 종속 변수 y(label)와 한 개 이상의 독립 변수 X(특성, column)와의 선형 상관 관계를 모델링하는 회귀 분석 기법

#### 단순 회귀 분석
$$y=w_1x_1+b$$  
#### 다중 회귀 분석
$$y=w_1x_1+w_2x_2+...+w_ix_i+b$$  

연속형 레이블 값을 예측하는 것이 목적일 경우, 선형 회귀를 사용해 데이터에 적합한 예측 모델을 개발할 수 있음

Linear Regression의 구현 방법으로는 `최소제곱법`과 `경사하강법`이 존재

<hr>

최소제곱법과 경사하강법에 대해 알아보기 전에 선형회귀에서의 **`손실`**과 **`회귀 분석 성능 평가 척도`**에 대해 알아보자

#### 선형 회귀에서의 손실, 오차 원리

- 선형 회귀에서는 (예측값-실제값)을 loss(손실, 오차)라고 한다.
- 위 그림과 같이 loss가 1, 0, -1인 경우 손실의 총합은 0이 된다.
- 실제로는 손실이 존재하지만 손실의 총합만 보게 되면 오류가 없는 것으로 보인다.  
  
  
- 이를 해결하기 위해 절대값이나 제곱을 사용하는 R-Squared, MAE, MSE, RMSE와 같은 방법으로 loss를 계산하여 성능을 평가한다.

### 회귀 분석 성능 평가 척도
- **R-Squared**
    - 실제 값의 분산 대비 예측 값의 분산 비율
    - 최고 : 1, 최악 : 음수
    - $$R^2 = 1-\frac{\sum_{}(y-\hat y)^2}{\sum_{}(y-\bar y)^2}$$
- **MAE(Mean Absolute Error)**
    - 실제 값과 예측 값의 차이를 절댓값으로 변환해 평균 계산
    - $$MAE = \frac{1}{m}\sum_{i=1}^m |y-\hat y|$$
- **Mean Squared Error**
    - 실제 값과 예측 값의 차이를 제곱해서 평균 계산
    - 이상치에 민감함
    - $$MSE = \frac{1}{m}\sum_{i=1}^m (y-\hat y)^2$$
- **RMSE**
    - 모델간의 잔차를 비교할 수 있음
    - 이상치에 덜 민감함
    - $$MSE = \sqrt{\frac{1}{m}\sum_{i=1}^m (y-\hat y)^2}$$
